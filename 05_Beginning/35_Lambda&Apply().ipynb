# 1 lambda Function

```python
# basic function
def sum_value(a,b):
    x= a+b
    return x
```

```python
sum_value(3,5)
```

```python
# lambda function (one liner)
sum_value_lambda = lambda a,b: a+b
```

```python
sum_value_lambda
```

```python
sum_value_lambda(4.5,7.3)
```

### Data NEw dataframe of dOb

```python
# Convert the 'DOB' column to datetime format
df["DOB"] = df["DOB"].astype("datetime64[ns]")

# # use lambda function to get the year or "dob" series
df["DOB"].apply(lambda x:x.year)
```

```python
# OR safer way (recommended)
df["DOB"] = pd.to_datetime(df["DOB"], errors='coerce')

# shows series by lambda function
df["DOB"].apply(lambda x: x.year)
```

```python
# Now extract the year using lambda
df["Year"] = df["DOB"].apply(lambda x: x.year)
```
