```python
import pandas as pd
```

```python
# reads csv file
df = pd.read_csv("InternationalCricketor.csv")
# set index
df.set_index("Player Name", inplace=True)
# select Column
df = df[["Matches",	"Innings",	"Runs",	"Wickets"]]
df
```

# 1 Drop Row

```python
# drop one row 
# axis parameter
df.drop("Rohit Sharma", axis=0)  # 0-> row  & 1-> column
```

```python
# index parameter
df.drop(index=["Rohit Sharma"])
```

```python
# drop two or more rows and update data (inplace=True)
df.drop(["Virat Kohli", "Rohit Sharma"], axis=0, inplace=True)
```

```python
df
```

# 2 Drop Columns

```python
# drop one column
# axis parameter
df.drop("Matches", axis=1)
```

```python
# column parameter
df.drop(columns=["Matches"])
```

```python
df.drop(columns=["Matches","Runs"])
```

```python
# drop column by position/index (drop last column)
df.columns[[1]]
```

```python
df.columns[[2]]
```

```python
# drop these column
df.drop(df.columns[[-1]], axis=1)
```

```python
# drop two ro more columns and update the data (inplace=True)
df.drop(["Matches", "Runs"], axis=1, inplace=True)
```
