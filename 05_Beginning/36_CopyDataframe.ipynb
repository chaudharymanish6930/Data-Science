```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv("CricketPlayers.csv")
# set index 
df.set_index("Name", inplace=True)
# select column
df = df[["Age", "DOB", 'Height (m)', "Weight (kg)", "Runs", "Country", "Club"]]
# show the dataframe
df.head(20)
```

# 1 copy()

## 1.1 deep = True

```python
# deep=Trur by default (modification to the data or indices of the copy)
df_copy = df.copy()
```

```python
# update value in original  dataframe
df.loc["David Roy","Height (m)"]
```

```python
# update value in original  dataframe
df.loc["David Roy","Height (m)"] = 22
```

```python
# copy vs original dataframe
df
```

## 1.2 deep = True

```python
# deep=false (any changes  to the data of the original will be reflected in the shallow copy and vice versa)
df_copy = df.copy(deep=False)
```

```python
# original the value in original dataframe
df.loc["David Roy", "Weight (kg)"]
```

```python
# update the original values
df.loc["David Roy", "Weight (kg)"]=100
```

```python
# shallow vs original copy
df
```

```python
# shallow copy
df_copy
```
