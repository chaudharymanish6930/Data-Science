## drop_duplicate()

```python
import pandas as pd
```

```python
df= pd.read_csv("LaptopData.csv")
```

```python
df
```

```python
df.head(10)
```

```python
# drop duplicates in 2 or more columns'
df.drop_duplicates(["Brand"])
```

```python
# drop duplicates and return the non-duplicates in 2 or more columns'
df.drop_duplicates(["Brand"])[["Brand","RAM_GB"]]
```

```python
# count the values
```

```python
df.drop_duplicates(["Brand"])[["Brand","RAM_GB"]].value_counts()
```

```python
df.drop_duplicates(["Brand"])[["Brand","RAM_GB"]].value_counts("Brand")
```

```python
# sort dataframe ascending by company and price
# cheapest first and most expensive last
df.sort_values(["Brand","Price_USD"])
```

```python
## assign or update the dataframe on same
df = df.sort_values(["Brand","Price_USD"])
```

```python
# cheapest : keep="First"
df.drop_duplicates(["Brand"], keep="first")
```

```python
# cheapest : keep="First"
df.drop_duplicates(["Brand"], keep="first")[["Brand","Price_USD"]]
```

```python
# most expensive: keep="last"
df.drop_duplicates(["Brand"], keep="last")[["Brand","Price_USD"]]
```

```python
df.drop_duplicates(["Brand"], keep="last", inplace=True)
```

```python
df
```

```python
df.drop_duplicates(["Brand"], keep="last", inplace=False)
```

```python
df.drop_duplicates(["Brand"], keep="last", inplace=True, ignore_index=True)
```

```python
df
```

```python
df.drop_duplicates(["Brand"], keep="last", inplace=True, ignore_index=False)
```
