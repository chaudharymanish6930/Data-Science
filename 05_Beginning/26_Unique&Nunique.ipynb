# unique()
##### Return unique values of series Object. (Uniques are returned in order of appearance)

```python
# get unique elements in the "brand" column
df["Brand"].unique()
```

```python
df["Screen_Size_inch"].unique()
```

```python
# get number of elements in the "brand" column
df["Brand"].nunique()
```

```python
# get number of unique elelments inn the "inches" column
df["Screen_Size_inch"].nunique()
```
