```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv("InternationalCricketor.csv")
# set index 
df.set_index("Player Name", inplace=True)
# select columns
df = df[["Country", "Matches", "Runs", "Wickets", "Innings"]]
df
```

# 1 Sample()

```python
# etract 10 random elements from the "Country" column
df["Country"].sample(10)
```

```python
df["Country"].sample(10, random_state=99)
# random state have random players
```

```python
df["Country"].sample(10, random_state=9)
# random state have random players
```

```python
# extract a random 20% sample of the dataframe
df.sample(frac=0.2)
```

```python
df.sample(frac=0.2, random_state=12)
```

```python
# upsample: increase the sampling rate -> frac>1
# note: replace parameter has to be True for frac parameter>1)
df.sample(frac=2, replace=True)
```

```python
df.sample(frac=2, replace=True, random_state=1)
```
