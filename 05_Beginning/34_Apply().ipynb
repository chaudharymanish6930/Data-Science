```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv("CricketPlayers.csv")
# set index
df.set_index("Name", inplace=True)
# select column
df= df[["Age", 'Height (m)', "Weight (kg)", "Runs", "Country", "Club"]]
# show dataframe
df
```

```python
# use numpy function and apply it to series
import numpy as np
df['Age'].apply(np.sqrt)
```

```python
# create your own funtcion and apply it to a dataframe
  # BMI = kg/m2
def calculate_BMI(row):
    return row['Weight (kg)']/( row["Height (m)"])**2
```

```python
df.apply(calculate_BMI, axis=1)
```
