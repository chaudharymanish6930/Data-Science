```python
import pandas as pd
```

```python
df_laptop = pd.read_csv("laptop_price.csv")
```

```python
df_laptop
```

#  1 isin(): Single Filtering

```python
# select Apple or HP laptops
df_laptop["Company"].isin(["Apple","HP"])
```

```python
df_laptop[df_laptop["Company"].isin(["Apple","HP"])]
```

```python
df_laptop[df_laptop["Company"].isin(["Apple","HP"])].value_counts()
```

```python
df_laptop[df_laptop["Company"].isin(["Apple","HP"])].value_counts("Company")
```

# 2 isin(): Multiple Filtering

```python
# find integrated or NVIDIA RTX 4060 from Apple and HP
filter1 = df_laptop["Graphics"].isin(["Integrated", "NVIDIA RTX 4060"])
filter2 = df_laptop["Company"].isin(["Apple","HP"])
```

```python
# filter dataframe
df_laptop[filter1 & filter2]
```
