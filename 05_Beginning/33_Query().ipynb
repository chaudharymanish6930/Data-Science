```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv("InternationalCricketor.csv")
# set index 
df.set_index("Player Name", inplace=True)
# select columns
df = df[["Country", "Matches", "Runs", "Wickets", "Innings"]]
df
```

# 1 query()

```python
# select players older than 34
df.query("Runs>9000")
```

```python
# exercise: write the quivalent boolean slicing
df[df["Runs"]>9000]
```

```python
# select players run more than 10000 from india
df.query("Runs>10000 and Country=='India'")
```

```python
# boolean slicing
df[(df["Runs"] > 10000) & (df["Country"] == "India")]
```

```python
# add not operator to the first example
df.query("not(Runs>9000)")
```

```python
# exercise: Write the equivalent boolean slicing
df[~(df["Runs"] > 9000)]
```

```python
# this is used to find run and above the 10000
df.query("Runs/10 > 1000")	
```

```python
# this is also boolean slicing
# df[(df["Runs"/10]>1000)]
df[(df["Runs"] / 10) > 1000]
```
