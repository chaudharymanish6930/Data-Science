### Task: Find the cheapest amd most expensive laptop of each company using the sort_values() and dulpicated() Methods (keep "First", "last" and false)

```python
import pandas as pd
```

```python
df= pd.read_csv("LaptopData.csv")
```

```python
df
```

```python
# sort the dataframe ascending by company amd price
# cheapest First and most Expensive at the end
df[["Brand","Price_USD"]].sort_values(ascending=False)
```

```python
df.sort_values(["Brand","Price_USD"],ascending=False)
```

```python
df.sort_values(["Brand","Price_USD"])
```

```python
df = df.sort_values(["Brand","Price_USD"])
```

```python
df
```

```python
df.value_counts("Brand")
```

```python
# duplicated values in "Company" column (default -> keep='first')
duplicated_first = df.duplicated("Brand", keep="first")
```

```python
# show dataframe with duplicated values in "company" column
df[duplicated_first]
```

```python
# keep = "first" ("vheapest laptop per company")
# show dataframe without duplicated values in "company" column
df[~duplicated_first]
```

```python
df[~duplicated_first][["Brand", "Price_USD"]]
```

```python
# check all vlaue categories
df[~duplicated_first].value_counts("Brand")
```
