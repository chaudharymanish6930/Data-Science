```python
import pandas as pd
df = pd.read_csv('InternationalCricketor.csv')
df.head(10)
```

```python
# set Index
df.set_index("Player Name", inplace=True)
```

##  1 Selecting With a single value
##### loc[row_label, column_label]

```python
# get all data about Rohit Sharma 
df.loc['Babar Azam']	
```

```python
# get the runs of babar azam
df.loc['Babar Azam',"Runs"]
```

```python
# get the average of the rohit sharma
df.loc["Rohit Sharma","Average"]
```

```python
# get all rows inside the "Matches" column
df.loc[:, "Matches"]
```

```python
# get all the columns that corespond to the index "Virat Kholi"
df.loc["Virat Kohli",:]
```

# 2 Selecting with a list of Values

```python
# get all data about Rohit Sharma and Virat Kohli
# by the list
# ["Virat Kohli", "Rohit Sharma"]

df.loc[["Virat Kohli", "Rohit Sharma"]]
```

```python
# get the matches of thee "["Virat Kohli", "Rohit Sharma"]"
df.loc[["Virat Kohli", "Rohit Sharma"], "Matches"]
```

```python
# get the matchhes and runs of Rohit Sharma
# ["Matches","Runs"]
df.loc["Rohit Sharma",["Matches","Runs"]]
```

```python
# get Matches and runs Of VIrat kohli and Rohit Sharma
df.loc[["Rohit Sharma","Virat Kohli"],["Matches","Runs"]]
```

# 4 Selecting with Conditions

```python
# one condition: select player with Strike rate above 90.00 
columns = ["Matches", "Runs", "Strike Rate", "Wickets",	"Bowling Average"]

df.loc[df["Strike Rate"]>90, columns]
```

```python
# multiple condtions : select player with Strike Rate above 90 and matches above 200
df.loc[(df["Strike Rate"]>90) & (df["Matches"]>200), :]
```

```python
# multiple condtions : select player with Strike Rate above 90 and matches above 200
df.loc[(df["Strike Rate"]>90) & (df["Matches"]>200), :"Strike Rate"]
```
