# 4 Selecting with option
#### iloc cannot acecept a boolean series, but only a bollean list. we have to use the list() function to convert a series into a boolean list

```python
# one condition: select player with runs above 10000
# columns = ["Innings	Runs	Average	Strike Rate	Wickets"]
columns =[1,2,3,4]
#  liat(df["Runs"]>10000)

df.iloc[list(df["Runs"]>10000), columns]
```

```python
# multiple conditions: select player with run amd country
# list((df["Runs"]>10000) & (df["Country"]=="Australia"))

df.iloc[list((df["Runs"]>5000) & (df["Average"]>50)),:]
```
