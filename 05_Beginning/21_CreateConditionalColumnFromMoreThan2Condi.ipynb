```python
import pandas as pd
```

```python
df = pd.read_csv("LaptopData.csv")
```

```python
df
```

```python
import numpy as np
```

```python
# convert the string type into the int type
df["Price_USD"] = pd.to_numeric(df["Price_USD"], errors="coerce")
```

```python
# Define conditions
conditions = [
    df["Price_USD"] > 3000,
    (df["Price_USD"] > 2000) & (df["Price_USD"] <= 3000),
    (df["Price_USD"] > 800) & (df["Price_USD"] <= 2000),
    df["Price_USD"] <= 800
]

# Define values
values = ["Too Expensive", "Expensive", "Affordable", "Cheap"]

# Apply np.select
df["Price_Category"] = np.select(conditions, values, default="Unknown")
```

```python
# shows the dataframe
df
```

```python
df["Price_Category1"] = np.select(conditions, values, default="Unknown")
```
