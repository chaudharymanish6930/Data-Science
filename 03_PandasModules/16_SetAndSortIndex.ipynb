# 1 Create Index

```python
import numpy as np
import random
```

```python
np.arange(1,100)
```

```python
# Create non-repeating values for the index
new_index = np.arange(1,100)
```

```python
new_index
```

```python
# Shuffling the indexes

random.shuffle(new_index)
```

```python
new_index
```

```python
import pandas as pd
df =pd.read_csv("students.csv")
df
```

```python
my_index = np.arange(1,14)
random.shuffle(my_index)
my_index
```

```python
df["my_index"]=my_index
```

```python
df
```

# 2 Set index

```python
df.set_index('my_index')
```

```python
df.set_index('my_index', inplace=True)
df
```

# 3 Sort Index

```python
# ascending order
df.sort_index()
```

```python
# decreasing order of the index
df.sort_index(ascending=False)
```

```python
df.sort_index(ascending=False, inplace=True)
```
