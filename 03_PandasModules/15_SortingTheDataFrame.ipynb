# Sort a DataFrame

```python
# by default ascending order
df.sort_values("maths")
```

```python
# by deafult aces
df.sort_values(by="hindi")
```

```python
# descending order of one colunm
df.sort_values("hindi", ascending=False)
```

```python
#  mutliple rows descending order arrange
df.sort_values(["maths", "hindi"], ascending=False)
```
