```python
# create a list of league
leagues = ["E0", "E1", "E2", "E3", "EC"]
frames = []

# loop for, read multiplle csv and append it into a list
for league in leagues:
    df = pd.read_csv(root + "2526"+ "/" + league + ".csv", encoding='unicode_escape')
    frames.append(df)
```

```python
# create a list of league
leagues = ["E0", "E2", "E3"]
frames = []

# loop through range of seasons and insert new columnss
for league in leagues:
    for season in range(15,21):
        df = pd.read_csv(root + str(season)+ str(season+1) + "/" + league + ".csv")
        df.insert(1,"Season",season)
        frames.append(df)
```

# 2.3 Organise the data in the dictionary

```python
# craete a dictionary with original name of league as key
dict_countries={
    "Spanish La Liga": "SP1", "Spanish Segunda Division": "SP2", "German Bundesliga":"D1",
"English Premier League":"E0", "English League 1":"E2", "English League 2":"E3" 
}
```

```python
# get or access the dictionary elelment
```

```python
dict_countries["Spanish La Liga"]
```

```python
# access all keys
for league in dict_countries:
    print(league)
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data={}

# looping through key elelments
for league in dict_countries:
    frames =[]
    for season in range(15,21):
        df=pd.read_csv(root + str(season) + str(season+1) + "/" + dict_countries[league] + ".csv")
        df.insert(1, "season", season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```
